```python
from tkinter import *
import random

from palabras import palabras

root = Tk()
root.title("Juego del Ahorcado")
root.geometry("320x240")

def obtenerPalabra():
    palabra = random.choice(palabras)
    return palabra


def intentoLetra(*args):
    try:
        letra = entrada.get()
        letra.upper()
        if letra in palabra:
            aciertos.append(letra)
    except ValueError:
        lresultado.set("Error")
            
        
    
    
# Marcos

marco1 = LabelFrame(root, pady = 3, padx = 12)
marco1.grid(row = 0, column = 0)

marcoIntentos = LabelFrame(root, pady = 7, padx = 12)
marcoIntentos.grid(row = 1, column = 0)

# Variables

abecedario = "ABCDEFGHIJKLMNÃ‘OPQRSTUVWXYZ"

palabra = obtenerPalabra()

letras_por_adivinar = set(palabra)

intentos = len(palabra) * 2

aciertos = []

letra = StringVar()

# Etiquetas

label1 = Label(marco1, text = f"palabra: {palabra}") # Etiqueta provisoria que muestra la palabra
label1.grid(row = 0, column = 0)

pal_lista = [letra if letra in aciertos else '_' for letra in palabra]

label2 = Label(marco1, text = pal_lista)
label2.grid(row = 1, column = 0)

lIntentos = Label(marcoIntentos, text = f"Intentos restantes: {intentos}")
lIntentos.grid(row = 0, column = 0)

tEntrada = Label(marcoIntentos, text = "Ingresa letra")
tEntrada.grid(row = 2, column = 0)

entrada = Entry(marcoIntentos, width = 2, textvariable = letra)
entrada.grid(row = 2, column = 1)
entrada.focus()

resultado = StringVar()
lresultado = Label(marcoIntentos, textvariable = resultado)
lresultado.grid(row = 3, column = 1)

btn1 = Button(marcoIntentos, text = "Ingresar", command = lambda: intentoLetra(letra))
btn1.grid(row = 2, column = 2)







root.bind("<Return>", intentoLetra)
root.mainloop()
```
